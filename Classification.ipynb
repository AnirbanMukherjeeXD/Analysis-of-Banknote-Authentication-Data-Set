# Import Libraries





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

# Load Dataset


```python
dataset = pd.read_csv("C:/Users/Kingshuk/Desktop/dataset.csv")
dataset.head()


```

```python
# Feature and Label
features = dataset.iloc[:, 0:4]
label = dataset.iloc[:, 4]
```

```python
plt.plot(dataset)
```

# Correlation Matrix

```python
plt.matshow(dataset.corr())
```

# K Nearest Neighbour


## Generate Model

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3)


from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier(n_neighbors=100)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

```

## Accuracy

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

# Support Vector Machine


## Generating Model

```python

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3) # 70% training and 30% test



from sklearn import svm

clf = svm.SVC(kernel='linear') 

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)


```

## Accuracy

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

# Random Forest

## Generating Model

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3) # 70% training and 30% test


from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=100)


clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)

```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
